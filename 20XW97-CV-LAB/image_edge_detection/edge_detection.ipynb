```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros([256, 256])
```

```python
img[0:256, 0:128] = 255
```

```python
plt.imshow(img, cmap='gray')
```

```python
# Passing the test image to mask. [Prewitt]
gx = np.array([[-1, 0, 1],[-1, 0, 1],[-1, 0, 1]])
gy = gx.T
```

```python
GX = cv2.filter2D(img, -1, gx)
GY = cv2.filter2D(img, -1, gy)
```

```python
G = np.sqrt(GX**2+GY**2)
A = np.arctan2(GY, GX)
```

```python
plt.figure(1)
plt.imshow(img, cmap='gray')
plt.title('Input Image')
plt.figure(2)
plt.subplot(2, 2, 1), plt.imshow(GX, cmap='gray')
plt.subplot(2, 2, 2), plt.imshow(GY, cmap='gray')
plt.subplot(2, 2, 3), plt.imshow(G, cmap='gray')
plt.subplot(2, 2, 4), plt.imshow(A, cmap='gray')
```

```python
img = np.zeros([256, 256])
for i in range(256):
    for j in range(i):
        img[i][j] = 255
```

```python
plt.imshow(img, cmap='gray')
```

```python
gx = np.array([[-1,0,1], [-1, 0, -1], [-1, 0, 1]])
gy = gx.T
GX = cv2.filter2D(img, -1, gx)
GY = cv2.filter2D(img, -1, gy)
G = np.sqrt(GX**2+GY**2)
A = np.arctan2(GY, GX)
```

```python
plt.figure(1)
plt.imshow(img, cmap='gray')
plt.title('Input Image')
plt.figure(2)
plt.subplot(2, 2, 1), plt.imshow(GX, cmap='gray')
plt.subplot(2, 2, 2), plt.imshow(GY, cmap='gray')
plt.subplot(2, 2, 3), plt.imshow(G, cmap='gray')
plt.subplot(2, 2, 4), plt.imshow(A, cmap='gray')
```
