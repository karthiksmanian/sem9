```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
t=np.linspace(0,1,100)
x=signal.square(2*np.pi*5*t)
h=np.array([1,-1])
y=signal.convolve(x,h,'same')
plt.subplot(2,1,1),plt.plot(t,x)
plt.title('Input signal')
plt.subplot(2,1,2),plt.plot(t,y)
plt.title('Output signal')
plt.tight_layout()
```

```python
h=np.ones(10)*(1/10) # Low pass filter
y=signal.convolve(x,h,'same')
plt.subplot(2,1,1),plt.plot(t,x)
plt.title('Input signal')
plt.subplot(2,1,2),plt.plot(t,y)
plt.title('Output signal')
plt.tight_layout()
```

```python
h=np.ones(10)*(1/10)
w,H=signal.freqz(h,1) #Convert to freq.domain
plt.plot(w,np.abs(H)) #Plot the absolute value
```

```python
h=np.array([1,-1])
w,H=signal.freqz(h,1) #Convert to freq.domain
plt.plot(w,np.abs(H)) #Plot the absolute value
```
