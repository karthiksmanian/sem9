```python
# Quantization - Mapping of large set of values to a smaller set of values. 
# Adv : (Compression)
# Drawback : Uniqueness is lost
```

```python
# Quantization is irreversible, once mapped into a value cannot be reversed.
# 1. Dynamic Range - max-min.
# 2. bit budget - b
# 3. L (levels) - 2**b
# 4. S = DR/L -> Quant. step
```

```python
# 1. generate a sinewave
# 2. Design of quantizer
# 3. quantization
# 4. Visualization
```

```python
# Uniform Quantization
import numpy as np
import matplotlib.pyplot as plt
```

```python
# step1 - Generate a sinewave
t = np.linspace(0, 1, 100)
pi = np.pi
x = np.sin(2*pi*5*t)
```

```python
plt.plot(t, x)
```

```python
# step2 - Design of quantizer
dr = np.max(x)-np.min(x)
b = 1
L = 2**b
S = dr/L
```

```python
# step3 - Quantization
y = np.round(x/S)*S
```

```python
# step4 - Visualization
plt.plot(t, x, 'r', t, y, 'k')
plt.legend(['input', 'output'], loc=1)
```
