```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Checkerboard Pattern
```

```python
image = np.zeros([8, 8])
```

```python
flg = False
m, n = image.shape

for i in range(m):
    flg = not flg
    for j in range(n):
        image[i][j] = 255 if flg else 0
        flg = not flg

# image[0][1] = 128
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
# e^i(theta) = cos(theta) + i*sin(theta)
# fix theta = -pi => cos alternates bwin -1, 1
for i in range(m):
    for j in range(n):
        image[i][j] = (-1)**(i+j)
```

```python
plt.imshow(image, cmap='gray')
```

```python
plt.plot(image[3, :])
plt.plot(image[4, :])
```

```python
image = np.zeros([200, 200])
```

```python
counter = 1
toggle = False

m, n = image.shape
for i in range(m):
    if counter%10==0:
        toggle = not toggle
        
    for j in range(n):
        if counter%10==0:
            toggle = not toggle
        image[i][j] = 255 if toggle else 0
        counter += 1
```

```python
plt.imshow(image, cmap='gray')
```

```python
# bed of impulse/comb function
image = np.zeros([256, 256])
```

```python
image[0:256:10, 0:256:10] = 255        
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
# Transforms
# 1. Fourier Transform is a complex quantity
# 2. Fourier transform of a signal is called spectrum. (complex)
# spectrum is a complex no from which magnitude and phase can be extracted.
```

```python
# np.abs() - mag. np.angle() - phase.
# np.fft.ff2() -> fourier transform
```

```python
img_spectrum = np.fft.fft2(image)
```

```python
img_spectrum[0][0], type(img_spectrum[0][0])
```

```python
IMG = np.fft.fftshift(img_spectrum)
```

```python
plt.imshow(image, cmap='gray')
```

```python
plt.imshow(np.abs(IMG), cmap='gray')
```

```python
# 2D sinusoidal grating
```

```python
x = np.linspace(-128, 128, 256)
y = np.linspace(-128, 128, 256)

X, Y = np.meshgrid(x, y)
```

```python
fx, fy = 5, 0
```

```python
img_sin = np.sin(2*np.pi*(fx*X*2+fy*Y*2))
```

```python
plt.imshow(img_sin, cmap='gray')
```

```python
plt.plot(img_sin[127])
plt.plot(img_sin[128])
plt.plot(img_sin[129])
```

```python
img_cos = np.cos(2*np.pi*(fx*X+fy*Y))
```

```python
plt.imshow(img_cos, cmap='gray')
```

```python
plt.plot(img_cos[127])
plt.plot(img_cos[128])
plt.plot(img_cos[129])
```

```python
img_spectrum = np.fft.fft2(img_sin)
```

```python
IMG = np.fft.fftshift(img_spectrum)
```

```python
plt.imshow(np.abs(IMG), cmap='gray')
```
