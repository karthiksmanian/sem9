```python
# 2. Hearing a sound
```

```python
fs = 8000 # SamplingFrequency - Capturing 8k samples in 1 sec.
f1 = 1000 #SignalFrequency - 1Khz 1000 Osciallations in a sec.
f2 = 2000
f3 = 3000
```

```python
t = np.linspace(0, 1, fs)
```

```python
y1 = np.sin(2*pi*f1*t)
y2 = np.sin(2*pi*f2*t)
y3 = np.sin(2*pi*f3*t)
```

```python
y = np.concatenate([y1, y2, y3])
sd.play(y, fs)
```

```python
plt.plot(t[0:100], y2[0:100])
```

```python
y2 = np.sin(2*pi*f2*t**2)
```

```python
plt.plot(t[0:1000], y2[0:1000])
```

```python
# Aliasing
# Increasing the signal frequency step by step while having Fs constant.
# Fs = 8000 - f's upperlimit is 4000. (Fs>=2*fmax)
# Violating Shannon's Theorem.
```
