```python
# Generate an array of size 256x256 of size 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros([256, 256])
image[128][128] = 255 #setting mid value to white
```

```python
# 2D impulse Function
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
image[128, :] = 255 # setting mid row to white
image[:, 128] = 255 # setting mid col to white
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
plt.plot(image[128, :])
```

```python
for i in range(256):
    image[i][i] = 255
    image[i][255-i] = 255
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
image[118:138, 118:138] = 255
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
image = np.zeros([256, 256])
```

```python
for i in range(256):
    for j in range(256):
        if (i-128)**2+(j-128)**2<=1:
            image[i][j] = 255
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
for i in range(256):
    for j in range(256):
        if (i-128)**2/50**2+(j-128)**2/100**2<=1:
            image[i][j] = 255
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```
