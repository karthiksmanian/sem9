```python
image = np.zeros([256, 256])
r = 25

image[((X-64)**2 + (Y-64)**2) <= r**2] = 255
image[((X-64)**2 + (Y+64)**2) <= r**2] = 255
image[((X+64)**2 + (Y+64)**2) <= r**2] = 255
image[((X+64)**2 + (Y-64)**2) <= r**2] = 255
```

```python
plt.imshow(image, cmap='gray')
```

```python
image = np.zeros([256, 256])

x = y = np.linspace(-128, 128, 256)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
distance = np.sqrt(X**2 + Y**2)

for r in range(10, 101, 10):
    image[np.abs(distance - r) < 1] = 1
```

```python
plt.imshow(image, cmap='gray')
```

```python
slope = (255-10)/(255-15)
yintercept = 255-slope*255
line = slope*xc + yintercept
width = 2
line_mask = (yc>=line - width) & (yc<=line+width)
image[line_mask] = 255
```
