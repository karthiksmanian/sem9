```python
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import signal
```

```python
img = cv2.imread('Z:/20XW97-CV-LAB/assets/images/casablanca.tif', 0)
```

```python
m, n = img.shape
```

```python
noise_prob = 0.1
threshold = 1-noise_prob
img_noise = np.zeros([m, n])
img_noise
```

```python
for i in range(m):
    for j in range(n):
        random_val = np.random.random()
        
        if random_val<noise_prob:
            img_noise[i][j] = 0
        elif random_val>threshold:
            img_noise[i][j] = 255
        else:
            img_noise[i][j] = img[i][j]
```
